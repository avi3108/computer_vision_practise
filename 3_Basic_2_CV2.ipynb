```python
#mouse events :


import cv2
import numpy as np

def draw(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONUP:
        cv2.circle(img,(x,y),10,(0,255,0),5)
    if(event == cv2.EVENT_RBUTTONUP):
        cv2.rectangle(img,(x,y),(x+100,y+75),(0,0,255),8)

cv2.namedWindow(winname="res")

img = np.zeros([600,600,3],np.uint8)*255
cv2.setMouseCallback("res",draw)

while (True):
    cv2.imshow("res",img)
    if cv2.waitKey(1)& 0XFF == 27:
        break

cv2.destroyAllWindows()
```

```python
#my made pixel value and pixel color shower
import cv2 
import numpy as np

cv2.namedWindow(winname="res")
img = cv2.imread("images/cat.18.jpg")

def getPixAndColor(events,x,y,flag,parms):
    cords = str(x) +","+ str(y)
    font = cv2.FONT_HERSHEY_PLAIN
    if events == cv2.EVENT_LBUTTONDOWN:
        cv2.putText(img,cords,(x,y),font,4,(22,44,255),6)
        cv2.imshow("res",img)
    if events == cv2.EVENT_RBUTTONDOWN:
        b = [x,y,0] 
        g = [x,y,1]
        r = [x,y,2]
        colors = str(b) + "," + str(g) + "," + str(r)
        cv2.putText(img,colors,(x,y),font,4,(225,225,225),6) 
        cv2.imshow("res",img)

cv2.setMouseCallback('res',getPixAndColor) 

while True :
    cv2.imshow("res",img)
    if cv2.waitKey(0) & 0XFF == 27:
        break

cv2.destroyAllWindows()
#
```
